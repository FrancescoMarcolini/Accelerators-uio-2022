```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
k1=4.7133E-5 # A/V^2 
k2=62.337e9 #V/m
d=18.8e-6 #m
V= np.array([600, 625, 650, 675, 700, 725, 749, 774]) #V
I=np.array([1.53E-11, 5.07E-11, 1.47E-10, 3.96E-10, 8.57E-10, 1.69E-09, 3.42E-09, 5.60E-09])#A
plt.plot(V,I)
plt.xlabel("V [V]")
plt.ylabel("I [A]")
```

```python
y=np.log(I*(d/V)**2)
x=d/V
```

```python
plt.plot(x,y, 'o-',label="experimental data")

plt.xlabel("x [$m/V$]")
plt.ylabel("y [$log(Am^2/V^2))$]")
plt.title("Fowler-Nordheim plot")
reg=stats.linregress(x,y)
a=reg.intercept
b=reg.slope
beta=-k2/b
A=np.exp(a)/(k1*(k2/b)**2)
r=np.sqrt(A/np.pi) #radius
print(f"A={A}[m^2]")
print(f"beta={beta}")
print(f"r= {r} [m]")

#fit risultante
yf=a+b*x
plt.plot(x,yf,label='linear fit')
plt.legend()

```
